### Loading Libraries

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import glob
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load Dataset Filenames

```python
#loading the names of the files as a list from the root folder
datasets = sorted(glob.glob('data*.csv'))
#printing the file name list 
print(datasets)
```

### Defining Necessary Functions

```python
#loading the dataset
def load_dataset(dataset):
    '''Returns DataFrame loaded from dataset'''
    df= pd.read_csv(dataset, header= None)
    return df

#describing the dataset with key statistics
def descriptive_stats(dataset,df):
    '''Prints descriptive statistics of the dataframe '''
    df= pd.read_csv(dataset, header= None)
    print('Descriptive Statistics of the Dataset: ', dataset)
    print(df.describe().T)
    
#identifying issues based on requirements document
def issue_identification(df):
    '''Prints identified issues based on if statement'''
    list_max = df.max() #returns the maximum of each feature of dataframe
    #check if feature#0=0 and feature#20=20
    if (list_max[0]==0.0) and (list_max[20]==20.0):
        print("I doubt the integrity of this dataset")
    else:
        count = 0
        for r in df.min():
            count+=r
        if count==0:
            print("why is this happening?")
        
        else:
            print("Dataset looks good")

#visualizing the avg, max, min of dataframe features
def visualize(df):
    '''Plots graph of average, maximum, minimum of features of DataFrame '''
    feature_avg = df.mean().tolist() #returns the average of each feature of dataframe
    feature_max = df.max().tolist() #returns the maximum of each feature of dataframe
    feature_min = df.min().tolist() #returns the minimum of each feature of dataframe
    #using the above values to create a dataframe
    df_avg = pd.DataFrame(columns=['Features', 'Average', 'Maximum', 'Minimum'])
    df_avg['Features'] = df.columns
    df_avg['Average'] = feature_avg
    df_avg['Maximum'] = feature_max
    df_avg['Minimum'] = feature_min
    #plotting the average
    plt.plot(df_avg['Features'], df_avg['Average'],"-b", label="Average")
    #plotting the minimum
    plt.plot(df_avg['Features'], df_avg['Minimum'] ,"-r", label="Minimum")
    #plotting the maximum
    plt.plot(df_avg['Features'], df_avg['Maximum'] ,"-g", label="Maximum")
    plt.xlabel('Features')
    plt.ylabel('Values')
    #set location of legend
    plt.legend(loc="upper right")
    #plots the graph
    plt.show()
 
```

### Calling Functions for All Datasets

```python
#iterating over all files and making function calls
for dataset in datasets[:4]:
    #print current dataset name
    print('\n\nDataset Name: ',dataset)
    #loading the data into dataframe
    df = load_dataset(dataset)
    #getting descriptive statistics
    #descriptive_stats(dataset, df)
    #identifying issues in the dataset
    issue_identification(df)
    #visualizing dataset characteristics
    visualize(df)
```
